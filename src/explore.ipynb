## Identifico que mi variable objetivo es "Y", la cual me indica si el cliente acepta o no la suscripción

```python
df['y'].value_counts()
df['y'].value_counts(normalize=True)
```

```python
import matplotlib.pyplot as plt

proportion = df['y'].value_counts(normalize=True)

plt.bar(proportion.index, proportion.values)
plt.title("Distribución de la variable objetivo")
plt.ylabel("Proporción")

for i, v in enumerate(proportion.values):
    plt.text(i, v, f"{v*100:.1f}%", ha='center')

plt.show()
```

## Esto nos indica que el 88.7% de los clientes no contrata el servicio y sólo el 11.3% si lo hace. Ahora tengo que ver que variables son las que realmente necesito para mI EDA
